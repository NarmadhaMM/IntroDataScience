```
import pandas as pd
from pyspark.sql.types import StringType, IntegerType, FloatType, StructType, StructField
SegData = pd.read_csv('http://bit.ly/2P5gTw4')
SegData.head()
```

```
disease_dat = pd.read_csv('http://bit.ly/2KXb1Qi')
disease_dat.head()
```

```
rating = pd.read_csv('http://bit.ly/2TNQ6TK')
rating.head()
```

```
SegData.dtypes
```

```
schemaFields_SegData = StructType([StructField("age", IntegerType(), False),
                StructField("gender", StringType(), False),
                StructField("income", FloatType(), False),
                StructField("house", StringType(), False),
                StructField("store_exp", FloatType(), False),
                StructField("online_exp", FloatType(), False),
                StructField("store_trans", IntegerType(), False),
                StructField("online_trans", IntegerType(), False),
                StructField("Q1", IntegerType(), False),
                StructField("Q2", IntegerType(), False),
                StructField("Q3", IntegerType(), False),
                StructField("Q4", IntegerType(), False),
                StructField("Q5", IntegerType(), False),
                StructField("Q6", IntegerType(), False),
                StructField("Q7", IntegerType(), False),
                StructField("Q8", IntegerType(), False),
                StructField("Q9", IntegerType(), False),           
                StructField("Q10", IntegerType(), False),
                StructField("segment", StringType(), False)])
```

```
# create spark dataframe
SegData_df = spark.createDataFrame(SegData, schema = schemaFields_SegData)
# write data
SegData_df.write.format("parquet").mode("overwrite").save("/databricks/IDS/SegData_df")
```

```
# create spark dataframe
rating_df = spark.createDataFrame(rating)
# write data
rating_df.write.format("parquet").mode("overwrite").save("/databricks/IDS/rating_df")
```

```
# create spark dataframe
disease_dat_df = spark.createDataFrame(disease_dat)
# write data
disease_dat_df.write.format("parquet").mode("overwrite").save("/databricks/IDS/disease_dat_df")
```

```
%sql
/* Create database */
CREATE DATABASE IF NOT EXISTS IDS
```

```
%sql
/* create table in the database and point to the location */
drop table if exists IDS.SegData_df;
CREATE TABLE IDS.SegData_df
USING parquet
LOCATION '/databricks/IDS/SegData_df';
```

```
%sql
/* create table in the database and point to the location */
drop table if exists IDS.rating_df;
CREATE TABLE IDS.rating_df
USING parquet
LOCATION '/databricks/IDS/rating_df';
```

```
%sql
/* create table in the database and point to the location */
drop table if exists IDS.disease_dat_df;
CREATE TABLE IDS.disease_dat_df
USING parquet
LOCATION '/databricks/IDS/disease_dat_df';
```
